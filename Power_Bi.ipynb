```python
dataset.isnull().sum()
```

```python
# Dealing with missing values
dataset = dataset.fillna(0)
dataset.isnull().sum()
```

```python
# Take a Look to DataTypes
dataset.dtypes
```

```python
# Let's Change some data types to date
dataset['Opening date'] = pd.to_datetime(dataset['Opening date'],errors='coerce')
dataset['Closing submission date'] = pd.to_datetime(dataset['Closing submission date'],errors='coerce')
dataset['Closing validation date'] = pd.to_datetime(dataset['Closing validation date'],errors='coerce')
```

```python
# Let's get the data just to perform the gantt chart
df = dataset[['Area\u200b', '\xa0Technical Data\xa0Collection\u200b', 'Domain/business DC\u200b', 'Data Steward', 'Data Collection Type', 'New Operational Major update', 'Description','Opening date', 'Closing submission date', 'Closing validation date']]
df.head()
```

```python
# Let's put the data in the right format
df_gantt = pd.melt(df, 
                   id_vars=['Domain/business DC\u200b','Area\u200b', '\xa0Technical Data\xa0Collection\u200b','Data Steward','Data Collection Type', 'New Operational Major update', 'Description'], 
                   var_name='Stage', 
                   value_vars=['Closing submission date', 'Closing validation date'])

#df_gantt['Start Date'] = pd.concat([dataset['Opening date'] for _ in range(dataset.shape[1])], ignore_index=True)

df_gantt.head()
```

```python
# Let's Rename Some Columns TYo be Recognized on plotly
df_gantt.rename(columns={'Opening Date':'Start',
                         'value':'Finish',
                         'Stage':'Subtask',
                        'Domain/business DC\u200b':'Task'},inplace=True)
```

```python
df_gantt
```

```python
df_gantt['Start'] = pd.to_datetime(df_gantt['Start'],errors='coerce')
df_gantt['Finish'] = pd.to_datetime(df_gantt['Finish'],errors='coerce')
```

```python
#df_gantt.to_excel('gantt_data.xlsx')
```

```python
fig = ff.create_gantt(df_gantt,index_col='Subtask',
                      show_colorbar=True, bar_width=0.2, showgrid_x=True, showgrid_y=True, group_tasks=True)


fig.show()
```

```python
fig = px.timeline(df_gantt, x_start="Start", x_end="Finish", y="Resource", color="Resource")
```
